```python
# 여러개의 컬럼을 불러올때는?
# 불러오고자 하는 column들을 list로 만들어서 불러온다.

titanic[['age','survived','fare']]
```

```python
# index를 이용한 slicing

titanic[1:10]
titanic[1:40:5]   # 5 만큼 건너뛰기.
```

```python
# loc method : 명칭기반 추출.
# 숫자를 넣으면 추출하고자 하는 index를 기준으로 데이터 선택.

titanic.loc[[3,10,15]]
```

```python
# slicing한데, 끝자리 index까지 포함한다.

titanic.loc[2:7]
#titanic[2:7]
```

```python
# column을 기준으로 추출하려면?
# ','로 index와 column을 구분.

titanic.loc[:,['age','class']]
```

```python
# iloc method : 위치기반 추출

titanic.iloc[2:7]
```

```python
titanic.iloc[:,[0,6]]
```

```python
# 범위가 2개면 index와 column 순서로 추출해준다.

titanic.iloc[1:4, 0:3]
```

```python
# boolean indexing!
# bool 데이터 : True 혹은 False

titanic['age'] >= 30
```

```python
# 나이가 30이상인 승객들의 모든 정보를 추출하라!

titanic[titanic['age'] >= 30]
```

```python
# 나이가 30이상인 사람들의 좌석등급 데이터

titanic[titanic['age'] >= 30]['class']
```

```python
# 생존한 승객들의 모든 정보를 추출하라!

titanic[titanic['survived'] == 1]
```

```python
# multiple boolean indexing

# and : &  (shift + 7)
# or  : |  (shift + \)

# 조건과 조건 사이에는 괄호가 필요하다.

# 나이가 30이상이고, 생존한 승객들의 정보를 추출하라.

titanic[(titanic['age'] >= 30) & (titanic['survived'] == 1)]
```

```python
# 성별이 여자고, 탑승클래스가 1등급인 승객들의 모든 정보를 추출하라!


titanic[(titanic['sex'] == "female") & (titanic['pclass'] == 1)]
```

```python
# 우리가 필요한 데이터만 추출해보자.

df = titanic[['survived', 'pclass', 'sex', 'age', 'fare', 'alone', 'deck']]
df.head()
```

```python
# 결측치처리
# 결측치 여부 파악하기

df.isna()
```

```python
# True = 1, False = 0

df.isna().sum()
```

```python
df.drop('deck', axis = 1)
```

```python
# deck이 drop 안되었다!

df
```

```python
# 덮어쓰기 기능!

df.drop('deck', axis = 1, inplace = True)
```

```python
df
```

```python
# dropna : 결측치가 포함된 모든 레이블을 제거
# axis를 지정하지 않으면 기본적으로 axis = 0

df.dropna(inplace = True)
df
```

```python
df.count()
```

```python
# 결측치를 제거하는 이유??

# 1. 보통 결측치 자체가 노이즈.... 활용할 수 없는 데이터!
# 2. 결측치가 포함되어있으면 통계를 낼 수 없거나, 시각화 기능이 안될 수 있다.
```

```python
# 자료 요약하기
# describe의 기본은 숫자형자료.

df.describe()
```

```python
# 모든 데이터를 요약하고 싶다.

df.describe(include = 'all')
```

```python
# 범주형 자료만 보고싶다.

df.describe(include = 'object')
# df.describe(include = 'number')
```

```python
# 개별적인통계 산출

df.mean()
df.std()
df.max()
df.min()
df.median() # 중앙값
df.mode()   # 최빈값
```

```python
# 남자승객과 여자승객의 평균나이 각각 구하기.

print("남자승객의 평균 연령 : {:1.2f}".format(    df[    df['sex'] == 'male'   ]['age'].mean()    )     )
print("여자승객의 평균 연령 : {:1.2f}".format( df[df['sex'] == 'female']['age'].mean()  )  )
```

```python
# 시각화 라이브러리

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Boxplot

sns.boxplot(x = 'sex', y = 'age', data = df)
```

```python
# hue : 한 단계를 더 구분해주는 기능

sns.boxplot(x = 'sex', y = 'age', hue = 'survived', data = df)
```

```python
# 히스토그램.
# 승객들의 연령을 히스토그램으로 시각화 하기.
# bins : 구간의 갯수

plt.hist(df['age'], bins = 8)

# 그래프 커스터마이징
plt.title("Titanic age histogram")
plt.xlabel("age")
plt.ylabel("People")
plt.show()
```

```python
sns.distplot(df['age'])
```

```python
# 막대그래프

sns.countplot(x = 'sex', data = df)
```

```python
# 성별에 따른 사망/생존 자의 수?

sns.countplot(x = 'sex', hue = 'survived', data = df)
```

```python
# 탑승 클래스에 따른 사망/생존 자의 수를 각각 나타내기.


sns.countplot(x = 'pclass', hue = 'survived', data = df)
```
